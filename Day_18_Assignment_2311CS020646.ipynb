```python
import pandas as pd

file_path = "/content/drive/MyDrive/Day_18_Tours_and_Travels.csv"
df = pd.read_csv(file_path)
df
```

```python
df['Review_Text'].fillna('No Review', inplace=True)
df['Rating'].fillna(df['Rating'].median(), inplace=True)
df['Customer_Age'].fillna(df['Customer_Age'].median(), inplace=True)
df
```

```python
df.drop_duplicates(subset=['Review_Text'], keep='first', inplace=True)
df
```

```python
print(df.columns)
```

```python
df['Destination'] = df['Destination'].str.strip().str.lower()
df
```

```python
import numpy as np

q1 = df['Package_Price'].quantile(0.25)
q3 = df['Package_Price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df = df[(df['Package_Price'] >= lower_bound) & (df['Package_Price'] <= upper_bound)]
df
```

```python
df = pd.get_dummies(df, columns=['Destination'])
df
```

```python
cleaned_file_path = "/content/drive/MyDrive/Cleaned_Tours_and_Travels.csv"
df.to_csv(cleaned_file_path, index=False)
```
